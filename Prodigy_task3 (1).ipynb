```python
high_corr_cols=['pdays','poutcome','deposit']
```

```python
df1=df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols,inplace=True,axis=1)
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()  # Create an instance of LabelEncoder
df_encoded = df1.apply(lb.fit_transform)  # Apply fit_transform to each column
df_encoded
```

```python
df_encoded['duration'].value_counts()
```

```python
x=df_encoded.drop('duration',axis=1)
y=df_encoded['duration']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
print(11162*0.25)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
    acc=accuracy_score(y_test,y_pred)
    print('Accuracy_score',acc)
    cm=confusion_matrix(y_test,y_pred)
    print('Confusion Matrics\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score=model.score(x_train,y_train)
    test_score=model.score(x_test,y_test)
    print('Training Score',train_score)
    print('Testing Score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x_test)
ypred_dt
```

```python
eval_model(y_test,ypred_dt)
```
